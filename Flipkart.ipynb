```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1"
content=requests.get(url)
soup=BeautifulSoup(content.text,'lxml')
```

```python
content
```

```python
name=[]
pname=soup.find_all('div',class_="_4rR01T")
for i in pname:
    name.append(i.text)
```

```python

pro=[]
ram=[]
op=[]
sto=[]
disp=[]
wrnty=[]

lists=soup.find_all('li',class_="rgWa7D")
for i in range(len(lists)):
    p=lists[i].text
    if ('Processor' in p ):
        pro.append(p)
    elif ('RAM' in p):
        ram.append(p)
    elif ('Operating' in p):
        op.append(p)
    elif ('SSD' in p and 'GB' in p) or ('HDD' in p and 'TB' in p) or ('SSD' in p and 'TB' in p):
        sto.append(p)
    elif ('Display' in p):
        disp.append(p)
    elif ('Warranty' in p or 'warranty' in p):
        wrnty.append(p)


    
```

```python
price=soup.find_all('div',class_="_30jeq3 _1_WHN1")
pri=[]
for i in range(24):
    pri.append(price[i].text)
```

```python
import pandas as pd
df=pd.DataFrame({'Description':name,'Processor':pro,'RAM':ram,'Operationg System':op,'Storage':sto,'Dispaly':disp,'Warranty':wrnty,'Price':pri})
```
